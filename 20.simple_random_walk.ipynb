Use the np.random module to draw 1000 coin flips at once. ste these to 1 and -1 and compute the cumulative sum.

```python
import numpy  as np
nsteps=1000
draws=np.random.randint(0,2,size=nsteps)
draws
```

```python
steps=np.where((draws>0), 1, -1)
```

```python
walk=steps.cumsum()

walk
```

```python
walk.shape
```

```python
walk.min()
```

```python
walk.max()
```

#### How long it took the random walk to get at least 10 steps away from origin


```python
(np.abs(walk)>=10).argmax()
```
